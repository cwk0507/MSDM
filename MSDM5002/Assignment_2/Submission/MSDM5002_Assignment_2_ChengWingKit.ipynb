We have two integer numbers. We know both numbers are larger than 1, and their sum is smaller than 100.  
Person A only knows the summation of these two numbers, and person B only knows the
product of these two numbers. Then, below is a very interesting conversion between
A and B (both are telling the truth).  
B: I don’t know the values of the two numbers.  
A: I know you don’t know the values of the two numbers.  
B: Now, I know the values of the two numbers.  
A: If you know, I also know the values of these two numbers now.  
What are the values of these two numbers?
You need to write a python code to search for and output the right results of this logical puzzle.

```python
# Your Code Here
# Create a list of possible combination in p_com, with each item being a list with two number [n1,n2]
p_com = []
for i in range(2,98):
    j = i
    while i+j <100:
        p_com.append([i,j])
        j+=1
print('Let n1 and n2 be the two target numbers.')
print('At first, there are {} different combinations of n1 and n2 satisfying n1,n2 >1 and n1+n2<100'.format(len(p_com)))
print('Suppose S is the sum that A is holding and P is the product that B is holding.')
# For each list in p_com, append the sum and product of n1 and n2 to the list so that the format become [n1,n2,n1+n2,n1*n2]
for com in p_com:
    com.append(com[0]+com[1])
    com.append(com[0]*com[1])
# B knows the product of the two numbers (P) but doesn't know the the values of the two numbers. 
# We can deduce that P is not a unique number among all possible products.
# Creat a dictionary p_count to store the count of each product in p_com, then b_product to store all possible number that P can be.
p_count={}
for com in p_com:
    if com[3] in p_count:
        p_count[com[3]]+=1
    else:
        p_count[com[3]]=1
b_product = [i for i in p_count if p_count[i]>1]
print("Since B doesn't know the values of n1 and n2, P should not be a unique number among all possible products.")
print("So, there are {} posible products that B may hold.".format(len(b_product)))
#print(b_product)
# A knows the sum of the two numbers (S) and he knows B doesn' t know the values of the two numbers.
# We can deduce that for every combinations that can add up to S, all products are not unique.
# Also, we assume that A doesn't know the two numbers at this stage, otherwise he can say he knows.
# So, 6 <= S <=99
# Create a_sum to store all possible number that S can be.
a_sum=[]
for i in range(6,100):
    i_sum_list = [com for com in p_com if com[2]==i]
    possible = True
    for ls in i_sum_list:
        if ls[3] not in b_product:
            possible = False
            break
    if possible:
        a_sum.append(i)
print("Since A knows B doesn't know n1 and n2, for every combinations that can add up to S, all the corresponding products are not unique.")
print("The possible S that A may hold are: {}".format(a_sum))
#print(a_sum)
# From A's message, B can deduce the list of possible S, i.e. a_sum
# B then knows the value of two numbers. We can deduce that among every possible n1, n2 such that n1*n2 = P, 
# there should be only one combination n1,n2 such that n1+n2 is in a_sum.
# update b_product as b_product_update 
b_product_update = []
for p in b_product:
    pp_list = [com for com in p_com if com[3]==p]
    pn = 0
    for ls in pp_list:
        if ls[2] in a_sum:
            pn+=1
    if pn==1:
        b_product_update.append(p)
print("After A claiming that he knows B doesn't know n1 and n2, B claims he knows n1 and n2.")
print("So, for every possible c1 and c2 such that c1*c2=P, there should be only one combination such that c1+c2 is in {}".format(a_sum))
print("Now, there are {} possible P that B may hold, namely PX: {}.".format(len(b_product_update),b_product_update))
# Finally, A also knows the value of two numbers. We can deduce that among every possible n1,n2 such that n1+n2 = S,
# there should be only one combination n1,n2 such that n1*n2 is in b_product_update.

a_sum_update = []
for s in a_sum:
    ss_list = [com for com in p_com if com[2]==s]
    sn=0
    for ls in ss_list:
        if ls[3] in b_product_update:
            sn+=1
    if sn==1:
        a_sum_update.append(s)
target = [com for com in p_com if (com[2]==a_sum_update[0]) and (com[3] in b_product_update)]

print("At last, A also knows n1 and n2 since B knows.")
print("So, for every possible c1 and c2 such that c1+c2=S, there should be only one combination such that c1*c2 is in PX.")
print("The only possible S is {}, and we can derive that n1 and n2 are {} and {}.".format(a_sum_update[0],target[0][0],target[0][1]))
```

Given a string, write `check(s)` to check if the opening brackets "(" and closing brackets ")" match each other and the number of matched pairs. Here we only consider this type of parenthesis pairs. However, do not use any recursive function. Your output should look like either of these:

Check parentheses: (a(b(c)))  
3 matched pairs  
True
***
Check parentheses: (a))(  
1 matched pairs  
False

```python
# Your Code Here
def check(s):
    op = 0
    pairs = 0
    correct = True
    for i in s:
        if i!='(' and i!=')':
            continue
        else:
            if i=='(':
                op+=1
            else: 
                if op>0:
                    pairs+=1
                    op-=1
                else: correct = False
    if pairs==1:
        print('1 matched pair')
    else:
        print('{} matched pairs'.format(pairs))
    return correct

s = input('Check parentheses: ')
print(check(s))


        
```

We assume a linked list node is defined as:

        class Node():
            def __init__(self, data):
                self.next = None
                self.data = data
Construct the input linked list and write a function `duplicate(head)` to duplicate a linked list (in-place). Here `head` is the id of the head node of a linked list. You should consider corner cases such as an empty linked list with `head = None`. Your output should look like this:

Enter a linked list: 0 -> 1 -> 2  
Duplicated linked list: 0 -> 1 -> 2 -> 0 -> 1 -> 2

```python
# Your Code Here
class Node():
    def __init__(self, data):
        self.next = None
        self.data = data
        
class L_List():
    def __init__(self):
        self.head = None
        self.tail = None
    def add_item(self,item):
        if item =='':
            return self
        # if item is not a Node, make it a Node
        if not isinstance(item, Node):
            item = Node(item)
        if self.head is None:
            self.head=item
        elif self.tail is not None:
            self.tail.next = item
        # update the tail
        t_tail = item
        while t_tail.next != None:
            t_tail = t_tail.next
        self.tail = t_tail
    def print_list(self):
        if self.head == None:
            print('Empty')
            return 
        else:
            active_node = self.head
            while active_node != None:
                print(active_node.data, end='')
                active_node = active_node.next
                if active_node != None:
                    print(' -> ', end='')
            print()

def duplicate(head):
    if head == None:
        return L_List()
    else:
        existing_list = L_List()
        existing_list.add_item(head)
        temp=L_List()
        active_node = head
        while active_node.next != None:
            temp.add_item(active_node.data)
            active_node = active_node.next
        temp.add_item(active_node.data)
        existing_list.tail.next = temp.head
        existing_list.tail = temp.tail
        return existing_list
    

input_list = input('Enter a linked list: ')
splited = input_list.split('->')
trimmed = [i.strip(' ') for i in splited]
l = L_List()
for item in trimmed:
    l.add_item(item)

print('Duplicated linked list: ', end='')
duplicate(l.head).print_list()
```

Write a function `revert(head)` to revert the linked list (in-place). Your output should look like this:

The reverted linked list is 2 -> 1 -> 0 -> 2 -> 1 -> 0.

```python
# Your Code Here
def revert(head):
    if head == None:
        return L_List()
    else:
        existing_list = L_List()
        existing_list.add_item(head)
        active_node = head
        previous_node = None
        while active_node.next != None:
            next_node = active_node.next
            active_node.next = previous_node
            if active_node == existing_list.head:
                existing_list.tail = active_node
            previous_node = active_node
            active_node = next_node
        active_node.next = previous_node
        existing_list.head = active_node
    return existing_list


r_list = revert(l.head)
print('The reverted linked list is ', end='')
r_list.print_list()
```

An airplane took off on 3 June 2020 at 9:24:51 pm GMT+9 and landed on 4 June 2020 at 12:27:42 am GMT+8 from a latitude and longitude of (35.769, 140.372) to (22.321, 113.928). Use the `datetime` module to calculate the flight duration. Print out the flight duration in hour:minute:second, straight-line distance in km and average speed in kph. Your output should look like this:

Duration:  
Distance:  
Speed:  

```python
# Your Code Here
import datetime
from math import sin, cos, sqrt, atan2, radians

start = datetime.datetime(2020,6,3,12,24,51)
end = datetime.datetime(2020,6,3,16,27,42)
duration = end - start
duration_in_hour = duration.seconds/3600
# approximate radius of earth in km
R = 6373.0

# Calculating distance
lat1 = radians(35.769)
lon1 = radians(140.372)
lat2 = radians(22.321)
lon2 = radians(113.928)
dlon = lon2 - lon1
dlat = lat2 - lat1
a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
c = 2 * atan2(sqrt(a), sqrt(1 - a))
distance = R * c

print('Duration: {}'.format(duration))
print('Distance: {} km'.format(distance))
print('Speed: {} kph'.format(distance/duration_in_hour))
```

```python
# Your Code Here

# The file containing the data is stored in the same directory of this assignment and named file.txt
from datetime import datetime,timedelta
import pandas as pd
import numpy as np
import math

# Read data 
file_path = './file.txt'
data = pd.read_csv(file_path,sep=" ",header=None)
data.columns = ['email','Join_date','Join_time','Leave_date','Leave_time']

# Format data
data['itcs']=data.email.str.extract('(\w+)@')
data['Join_dt']=data.Join_date.apply(lambda x: datetime(int(x[6:10]),int(x[3:5]),int(x[:2])))+data.Join_time.apply(lambda x: timedelta(int(x[:2])/24+int(x[3:5])/(24*60)+int(x[6:8])/(24*60*60)))
data['Leave_dt']=data.Leave_date.apply(lambda x: datetime(int(x[6:10]),int(x[3:5]),int(x[:2])))+data.Leave_time.apply(lambda x: timedelta(int(x[:2])/24+int(x[3:5])/(24*60)+int(x[6:8])/(24*60*60)))
data['Duration']=((data.Leave_dt-data.Join_dt)/np.timedelta64(1,'m')).apply(math.ceil)

# Define SUMIF function
def SUMIF(email,itsc,duration):
    td = 0
    for idx,val in enumerate(email):
        if val.split('@')[0]==itsc:
            td+=duration[idx]        
    return td

# Main 
itscs = input('Enter all the student itsc: ')
splited = sorted(itscs.split())
for i in splited:
    print('{:<10}{:>4}'.format(i,SUMIF(data.email,i,data.Duration)))
```
