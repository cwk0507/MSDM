```python
# Q2 answer
import math

def decimal_length(x):
    max_decimals = 16
    int_part = int(abs(x))
    frac_part = abs(x) - int_part
    multiplier = 10 ** max_decimals
    frac_digits = multiplier + int(multiplier * frac_part)
    while frac_digits % 10 == 0:
        frac_digits /= 10
    scale = int(math.log10(frac_digits))
    return scale 

def arange(start=0,stop=0,step=1):
    e = 1e-15
    target=[]
    if type(start)==float or type(stop)==float or type(step)==float:
        start = float(start)
        stop = float(stop)
        step = float(step)
        precision = max(decimal_length(start),decimal_length(stop),decimal_length(step))
        while start < stop - e:
            target.append(start)
            start = round(start + step,precision)
            if abs(start-int(start)) <e:
                start = float(int(start))
        if target==[]:
            if step > 0:
                step = -step
            while start > stop + e:
                target.append(start)
                start = round(start + step,precision)
                if abs(start-int(start)) <e:
                    start = float(int(start))
    else:
        while start < stop:
            target.append(start)
            start += step
        if target==[]:
            if step > 0:
                step = -step
            while start > stop:
                target.append(start)
                start += step
    return target

while True:
    a = input('Enter an expression with arange(): ')
    try:
        print(eval(a))
    except:
        break
```

3. An identity matrix is defined as a square matrix with 1s running from the top left of the square to the bottom right and the rest are 0s. Create a numpy.ndarray that takes an integer n and print the identity matrix of n x n dimensions. For this question, if the integer is negative, return the mirror image of the identity matrix of n x n dimensions. It does not matter if the mirror image is left-right or top-bottom. Incompatible types passed as n should return the string "Error". Your output should look like one of these:


    Enter dimension of identity matrix: 0
    []
***
    Enter dimension of identity matrix: 2
    [[1 0]
     [0 1]]
***
    Enter dimension of identity matrix: -2
    [[0 1]
     [1 0]]

```python
# Q3 answer
def identity(n):
    mirror = False
    if n<0:
        mirror = True
        n = -n
    matrix = np.zeros((n,n),int)
    if not mirror:
        one_position = 0
        r = 0
        while n>0:
            matrix[r,one_position]=1
            one_position += 1
            r += 1
            n -= 1
    else:
        one_position = n-1
        r = 0
        while n>0:
            matrix[r,one_position]=1
            one_position -= 1
            r += 1
            n -= 1
    return print(matrix)
    
n = input('Enter dimension of identity matrix: ')
try:
    identity(int(n))
except:
    print('Error')
```

5. Now add some AI to question 4 to override choosing a place randomly for each turn in the tic-tac-toe game. First, see if there is a place to take that will win the game. If there is, take that place. Otherwise, see if there is a place the other side can take that will win the game. If there is, take there to block the other side. Otherwise, continue with the randomly selected place remaining. Run the simulation until 1 second is reached. Your output should look like this:


    Probability of 1 wins: 
    Probability of 2 wins: 
    Probability of draw: 
    Games completed: 
    Simulation time: 

```python
# Q5 answer
def check_end(ttt):
    for i in range(3):
        if ttt[i,0] == ttt[i,1] and ttt[i,1] == ttt[i,2] and ttt[i,0]!=0:
            return True, ttt[i,0]
        if ttt[0,i] == ttt[1,i] and ttt[1,i] == ttt[2,i] and ttt[0,i]!=0:
            return True, ttt[0,i]
    if ttt[0,0] == ttt[1,1] and ttt[1,1] == ttt[2,2] and ttt[0,0]!=0:
        return True, ttt[0,0]
    if ttt[0,2] == ttt[1,1] and ttt[1,1] == ttt[2,0] and ttt[1,1]!=0:
        return True, ttt[1,1]
    if (ttt==0).sum() == 0:
        return True, 'Draw'
    return False, None

def check_win(ttt,player):
    for i in range(3):
        if (ttt[i,:]==player).sum()==2 and (ttt[i,:]==0).sum()==1:
            return True, (i,np.where(ttt[i,:]==0)[0][0])
        if (ttt[:,i]==player).sum()==2 and (ttt[:,i]==0).sum()==1:
            return True, (np.where(ttt[:,i]==0)[0][0],i)
    diagonal_1 = np.array([ttt[0,0],ttt[1,1],ttt[2,2]])
    diagonal_2 = np.array([ttt[0,2],ttt[1,1],ttt[2,0]])
    if (diagonal_1==player).sum()==2 and (diagonal_1==0).sum()==1:
        return True, (np.where(diagonal_1==0)[0][0],np.where(diagonal_1==0)[0][0])
    if (diagonal_2==player).sum()==2 and (diagonal_2==0).sum()==1:
        return True, (np.where(diagonal_2==0)[0][0],1-(np.where(diagonal_2==0)[0][0]-1))
    return False, None    
          
def TTT():
    TicTacToe = np.zeros((3,3),int)
    end = False
    player = ['',1,2]
    ap = 1
    turn = 1
    while not end and turn<=9:
        p_i = np.random.randint(0,3)
        p_j = np.random.randint(0,3)
        while TicTacToe[p_i,p_j] != 0:
            p_i = np.random.randint(0,3)
            p_j = np.random.randint(0,3)
        lose,p = check_win(TicTacToe,player[-ap])
        if lose:
            p_i = p[0]
            p_j = p[1]
        win,p = check_win(TicTacToe,player[ap])
        if win:
            p_i = p[0]
            p_j = p[1]
        TicTacToe[p_i,p_j] = player[ap]
        end, result = check_end(TicTacToe)
        if end:
            return result
        turn += 1
        ap = -ap
        
import time
games = 10000
tos = time.time()
results = {}
for i in range(games):
    r = TTT()
    if r not in results:
        results[r] = 1
    else:
        results[r]+=1

print('Probability of 1 wins: {}'.format(results[1]/games))
print('Probability of 2 wins: {}'.format(results[2]/games))
print('Probability of draw: {}'.format(results['Draw']/games))
print('Games completed: {}'.format(games))
print('Simulation time: {}s'.format(time.time()-tos))
```

6. A cereal company is giving away a prize in each box of cereal and they advertise "Collect all 10 prizes!". Each box of cereal has 1 prize, and each prize is equally likely to appear in any given box. Estimate how many boxes it takes, on average, to get N different prizes using simulation for just 0.1 second for each N. Your output should look like this:


    Total simulation time:
    
    N    Average boxes    
    1    1.0
    2

```python
# Your Code Below
import numpy as np
import time 
import timeit

def estimate(N):
    n = 1
    while True:
        prizes = np.random.randint(1,N+1,n)
        distinct_prizes = len(np.unique(prizes))
        if distinct_prizes == N:
            break
        n+=1
    return n


simulation_count = 10000
result = np.zeros([10,simulation_count],int)
tos = time.time()
for i in range(10):
    for j in range(simulation_count):
        result[i,j]=estimate(i+1)
print('Total simulation time: {}s'.format(time.time()-tos))
means = result.mean(axis=1)
print()
print('N    Average boxes    Average Simulation Time')
for i in range(10):
    print('{:<5}{:<17.2f}{:.6f}s'.format(i+1,means[i],timeit.timeit('estimate({})'.format(i+1), globals=globals(), number=simulation_count)/simulation_count))
```
