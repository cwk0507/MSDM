```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sqrt, pi, exp

def normal(x, mu, sigma):
    return 1/sqrt(2*pi*sigma**2)*exp(-(x-mu)**2/(2*sigma**2))

N=1000000
mu_x = 0
sigma2_x = 1
mu_y = 1
sigma2_y = 2
x = np.random.normal(mu_x, sqrt(sigma2_x), N)
y = np.random.normal(mu_y, sqrt(sigma2_y), N)
z = x + y
mu_z = np.mean(z)
sigma2_z = np.var(z)
print(f'mu_z = {mu_z:.2f}, sigma2_z = {sigma2_z:.2f}')

# Plot
plt.figure(figsize=(8,8))
plt.hist(x, 50, density=True, color='blue', label=f'X~N({mu_x},{sigma2_x})')
plt.hist(y, 50, density=True, color='orange', label=f'Y~N({mu_y},{sigma2_y})')
values, bins, _ = plt.hist(z, 50, density=True, color='green', label='Z=X+Y')
mu_z_theo = mu_x + mu_y
sigma2_z_theo = sigma2_x + sigma2_y
plt.plot(bins, normal(bins, mu_z_theo, sqrt(sigma2_z_theo)), linewidth=3, color='red', label=f'Theoretical Z~N({mu_z_theo},{sigma2_z_theo})')
plt.title(f'Density of {N:,} sampled X~N({mu_x},{sigma2_x}), Y~N({mu_y},{sigma2_y}) and Z=X+Y')
plt.legend()
```
