```python
import numpy as np
def U(x):
    return x**2/2

N = 1000000
x = np.zeros(N)
mu = 0
sigma = 1
steps = np.random.normal(mu, sigma, N-1)
for i in range(N-1):
    x_current = x[i]
    x_transition = x_current + steps[i]
    delta_E = U(x_transition) - U(x_current)
    r = min(np.exp(-delta_E),1)
    if np.random.uniform(0,1) <= r:
        x[i+1] = x_transition
    else:
        x[i+1] = x_current

print(f'<x> = {np.mean(x)}')
print(f'<x^2> = {np.mean(x**2)}')
print(f'<x^3> = {np.mean(x**3)}')
print(f'<x^4> = {np.mean(x**4)}')
```

```python
import matplotlib.pyplot as plt

def f(x):
    return (1/np.sqrt(2*np.pi))*np.exp(-x**2/2)

plt.hist(x, bins=100, density=True, label='x in Markov chain')
xf = np.linspace(min(x),max(x),10000)
plt.plot(xf,f(xf), label='f(x)')
plt.title('Density of x in Markov chain generated according to f(x)')
plt.xlabel('X')
plt.ylabel('Density')
plt.legend()
```
